```python
#This is me trying to teach myself tensorflow. Most of this code came from https://github.com/Hvass-Labs Thank you
#for the help
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from sklearn.metrics import confusion_matrix
from mnist import MNIST
data = MNIST(data_dir="data/MNIST/")
data.y_test[0:5, :]
# The images are stored in one-dimensional arrays of this length.
img_size_flat = data.img_size_flat

# Tuple with height and width of images used to reshape arrays.
img_shape = data.img_shape

# Number of classes, one class for each of 10 digits.
num_classes = data.num_classes

    
def inlineplot(images, cls_true, row = 3, column = 3, cls_pred=None):
    assert len(images) == row*column 
    assert len(cls_true) == row*column
    
    # Create figure with 3x3 sub-plots.
    fig, axes = plt.subplots(row, column)
    fig.subplots_adjust(hspace=0.3, wspace=0.3)

    for i, ax in enumerate(axes.flat):
        # Plot image.
        ax.imshow(images[i].reshape(img_shape), cmap='binary')

        # Show true and predicted classes.
        if cls_pred is None:
            xlabel = "True: {0}".format(cls_true[i])
        else:
            xlabel = "True: {0}, Pred: {1}".format(cls_true[i], cls_pred[i])

        ax.set_xlabel(xlabel)
        
        # Remove ticks from the plot.
        ax.set_xticks([])
        ax.set_yticks([])
        
    # Ensure the plot is shown correctly with multiple plots
    # in a single Notebook cell.
    plt.show()
    
def showimages(row, column):
    images = data.x_test[0:row*column]
    cls_true = data.y_test_cls[0:row*column]
    inlineplot(images=images, cls_true=cls_true, row = row, column = column)
```

```python
showimages(5,4)
```
